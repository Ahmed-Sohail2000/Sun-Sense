We are going to be using [DateTime](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html#pandas-to-datetime) method to convert the data into a format of day/month/year and having the datatype of datetime.

```python
# import datetime
from datetime import datetime

# convert the data columns of YEAR, MO, DY into a single column with the format of d/m/yr with the datatime datatype
data_df['YEAR'] = data_df['YEAR'].astype(str)
data_df['MO'] = data_df['MO'].astype(str).str.zfill(2) # fill the month column that can have upto 2 digits
data_df['DY'] = data_df['DY'].astype(str).str.zfill(2) # fill the day column to have upto 2 digits

# join the 3 columns into a single one
data_df['Date'] = data_df['DY'] + '/' + data_df['MO'] + '/' + data_df['YEAR']

# drop the YEAR, MO, DY columns and shift the Date column to the left
data_df.drop(columns = ['YEAR', 'MO', 'DY'], inplace = True)
data_df = data_df[['Date'] + [col for col in data_df.columns if col != 'Date']]

# convert the Date column into datetime type
data_df['Date'] = pd.to_datetime(data_df['Date'], format = '%d/%m/%Y')

# rename the columns by adding the units
data_df.rename(columns = {'ALLSKY_SFC_SW_DWN' : 'Solar Irradiance (kWh/m^2/day)',
                          'T2M_RANGE' : 'Temperature (2 meters)',
                          'WS10M_RANGE' : 'Wind Speed (10 meters)'},
               inplace = True)

# save the data_df
data_df.to_csv('Data/data_df.csv')
```

## 2.1 **Deep Learning Models ü§ñüñ•Ô∏èüë®üèª‚Äçüíª**


The model architecture that we would be implementing on our training data are listed below can be found using the [PyTorch Forecasting Models](https://pytorch-forecasting.readthedocs.io/en/stable/models.html#models)

1. Autoregressive Integrated Moving Average (ARIMA): ARIMA models are a classical approach for time series forecasting. They consider the autoregressive (AR), moving average (MA), and differencing (I) components of the time series to make predictions.

2. Long Short-Term Memory (LSTM) Networks: LSTM networks, a type of recurrent neural network (RNN), are effective for capturing long-term dependencies in time series data. They have a memory cell that can retain information over time and are suitable for modeling complex temporal patterns.

3. Gated Recurrent Unit (GRU) Networks: GRU networks are another type of RNN that can capture temporal dependencies in time series data. They have gating mechanisms that control the flow of information through the network and can be more computationally efficient than LSTMs.

4. Convolutional Neural Networks (CNNs): While CNNs are commonly used for image processing, they can also be applied to time series forecasting. By treating the time series as an image with one dimension, CNNs can extract features and learn patterns from the data.

5. Transformer Networks: Transformer networks have gained popularity in natural language processing tasks, but they can also be adapted for time series forecasting. Transformers leverage self-attention mechanisms to capture relationships between different time steps and have shown promising results in modeling sequential data.

6. Ensemble Models: Ensemble models combine multiple individual models to make predictions. For time series forecasting, ensemble models such as the combination of multiple ARIMA models or the combination of different LSTM models can improve overall accuracy and robustness.

It's important to experiment with [different architectures](https://pytorch-forecasting.readthedocs.io/en/stable/models.html#selecting-an-architecture) and select the one that best suits specific time series data and forecasting requirements. Additionally, [factors](https://pytorch-forecasting.readthedocs.io/en/stable/models.html#size-and-type-of-available-data) such as the complexity of your data, available computational resources, and interpretability of the model should also be considered when choosing an architecture.

How Encoder and Decoder actually works:

In the context of time series analysis and forecasting, the terms "encoder" and "decoder" are often associated with sequence-to-sequence models, which are a class of models used for tasks where the input and output are both sequences of data points, like time series forecasting. Here's a brief definition of these terms:

1. **Encoder:** The encoder is responsible for processing the input sequence (historical data) and transforming it into a fixed-size representation or context vector. It captures the relevant information from the past observations and summarizes it in a form that can be used by the decoder to make predictions. The encoder's output is typically used as the initial hidden state of the decoder.

2. **Decoder:** The decoder takes the fixed-size representation (context vector) from the encoder and generates the future sequence (forecasted values) step by step. It uses the context vector as well as the previous predictions (during training) or ground truth (during inference) to generate the next data points in the sequence. The decoder operates autoregressively, where each step generates one element of the output sequence at a time.

In time series forecasting, the encoder processes the historical observations (past time steps) and produces a context vector that encodes the relevant information. The decoder then uses this context vector to generate future predictions. The overall architecture is often referred to as the "encoder-decoder" architecture.

This type of architecture is commonly used in recurrent neural networks (RNNs), such as Long Short-Term Memory (LSTM) and Gated Recurrent Unit (GRU) networks, for time series forecasting tasks. It allows the model to capture complex temporal relationships and dependencies in the data, making it well-suited for predicting future values in a time series.

## 2.4 **Model 1: Temporal Fusion Transformer**

The first model that we will be training our dataset will be [TemporalFusionTransformer](https://pytorch-forecasting.readthedocs.io/en/stable/api/pytorch_forecasting.models.temporal_fusion_transformer.TemporalFusionTransformer.html#temporalfusiontransformer) model and we will be then finding the optimal learning rate for that model we create. After determining the optimal learning rate, we can then move on to training and evaluating the model to see how it performs.

```python
# import the necessary files
from pytorch_forecasting.models import TemporalFusionTransformer
import lightning.pytorch as pl
from lightning.pytorch import Trainer
from lightning.pytorch.tuner import Tuner
from pytorch_forecasting.metrics import QuantileLoss, MASE, SMAPE, RMSE

# configure the network and trainer
pl.seed_everything(42)
trainer = Trainer(accelerator = 'auto', gradient_clip_val = 0.1)

# create the model from the dataset
tft = TemporalFusionTransformer.from_dataset(
    train_dataset,
    learning_rate = 1,
    hidden_size = 8,  # most important hyperparameter apart from learning rate # number of attention heads. Set to up to 4 for large datasets
    attention_head_size = 1,
    dropout = 0.1,  # between 0.1 and 0.3 are good values
    hidden_continuous_size = 8,  # set to <= hidden_size
    loss = QuantileLoss(),
    optimizer = "Ranger"
)

# find the number of parameters in the model
print(f'Number of parameters in network: {tft.size()/1e3:.1f}k')
```

Explains what the code above does and means

1) train_dataset: This is the training dataset you've previously defined using TimeSeriesDataSet. It contains the necessary data, target, and features for training the model.

2) learning_rate: The learning rate controls the step size during gradient descent optimization. It affects the speed at which the model updates its weights during training.

3) hidden_size: The size of the hidden layer in the Transformer-based encoder and decoder blocks. It is an important hyperparameter that influences the model's capacity to learn complex patterns from the data.

4) attention_head_size: The size of the attention heads in the multi-head self-attention mechanism of the Transformer. It is usually set to 1 for most applications.

5) dropout: Dropout is a regularization technique used to prevent overfitting by randomly setting a fraction of the neurons to zero during training. The dropout rate here is set to 0.1, meaning 10% of neurons will be dropped out during training.

6) hidden_continuous_size: The size of the hidden layer in the continuous variables transformation block. It is set to 8, which should be less than or equal to the hidden_size.

7) loss: The loss function used for training the model. Here, the QuantileLoss() is used, which is suitable for quantile forecasting.

8) optimizer: The optimizer used during training. In this case, "Ranger" optimizer is used, which is a combination of RAdam and LookAhead optimizers.

### 2.4.1 **Finding The Optimal Learning Rate For Our Model**

Before starting to use any models, we are going to be finding the best learning rate for our data in order to train our data to various models that we will select and this can be done by using [Learning Rate Finder](https://lightning.ai/docs/pytorch/stable/api/lightning.pytorch.callbacks.LearningRateFinder.html) from PyTorch Lightning framework.

```python
# import the necessary libraries
import lightning.pytorch as pl
from lightning.pytorch.callbacks import EarlyStopping, LearningRateFinder, LearningRateMonitor
from lightning.pytorch.loggers import TensorBoardLogger
from lightning.pytorch import Trainer
from lightning.pytorch.tuner import Tuner

# Initialize the trainer with the necessary callbacks
pl.seed_everything(42)
trainer = pl.Trainer(
    logger = TensorBoardLogger("tb_logs", name="TFT"),
    accelerator = 'auto',
    gradient_clip_val = 0.1
)

# Find the optimal learning rate using lr_find from PyTorch Forecasting
lr_finder = Tuner(trainer).lr_find(tft, train_dataloaders = train_dataloader, val_dataloaders = valid_dataloader, min_lr = 1e-6, max_lr = 10)

# Print the suggested optimal learning rate
print(f"Suggested learning rate: {lr_finder.suggestion()}")
fig = lr_finder.plot(suggest=True)
fig.show()
```

```python
# import tensorflow and tensorboard, # import the necessary libraries
import tensorflow as tf
import tensorboard as tb
tf.io.gfile = tb.compat.tensorflow_stub.io.gfile
import lightning.pytorch as pl
from lightning.pytorch.callbacks import EarlyStopping, LearningRateMonitor
from lightning.pytorch.loggers import TensorBoardLogger
from lightning.pytorch import Trainer

# Load the TensorBoard notebook extension
%load_ext tensorboard

# configure network and trainer
pl.seed_everything(42)
early_stop_callback = EarlyStopping(monitor = 'val_loss', min_delta = 1e-4, patience = 10, verbose = False, mode = 'min')
lr_logger = LearningRateMonitor() # to log the learning rate
logger = TensorBoardLogger('lightning_logs') # logging results to tensorboard

# create a trainer
trainer = pl.Trainer(max_epochs = 10, accelerator = 'auto', enable_model_summary = True,
                     gradient_clip_val = 0.1,
                     limit_train_batches = 50,  # coment in for training, running valiation every 30 batches # fast_dev_run=True,  # comment in to check that networkor dataset has no serious bugs
                     callbacks = [lr_logger, early_stop_callback],
                     logger = logger)

tft = TemporalFusionTransformer.from_dataset(
    train_dataset,
    learning_rate = 0.06,
    hidden_size = 16,
    attention_head_size = 2,
    dropout = 0.1,
    hidden_continuous_size = 8,
    loss = QuantileLoss(),
    log_interval = 5,  # uncomment for learning rate finder and otherwise, e.g. to 10 for logging every 10 batches
    optimizer = "Ranger",
    reduce_on_plateau_patience = 4,
)

# print the number of parameters in the network
print(f"Number of parameters in network: {tft.size()/1e3:.1f}k")
```

5. `dropout = 0.1,`: This sets the dropout rate, which is a regularization technique used during training to prevent overfitting. It randomly sets a fraction of the model's input units to zero, helping to avoid relying too heavily on specific inputs.

6. `hidden_continuous_size = 8,`: This sets the number of hidden units used in the continuous feature transformation network. It is related to handling continuous features in the data.

7. `loss = QuantileLoss(),`: This sets the loss function used to train the model. The `QuantileLoss` is a type of loss function suitable for quantile-based forecasting, which estimates different quantiles of the target distribution.

8. `log_interval = 10,`: This parameter is used to control how often the training logs are printed during training. In this case, it is set to 10, which means the logs will be printed every 10 batches.

10. `reduce_on_plateau_patience = 4,`: This is a parameter related to a learning rate scheduler. It determines the number of epochs with no improvement in validation loss before reducing the learning rate during training.

These parameters are used to configure the `TemporalFusionTransformer` model for training and specify various hyperparameters that affect the model's architecture and training process.

Prophet has gained popularity in industries that require accurate and intuitive time series forecasting, such as retail, e-commerce, finance, and more.
```python
# import prophet
from prophet import Prophet

# create a separate sun power data from data_df
sun_power_data = data_df.copy()

# remove the Temperature, wind speed, and time idx
sun_power_data.drop(labels = ['Temperature (2 meters)', 'Wind Speed (10 meters)', 'time_idx'], axis = 1, inplace = True)

# rename the columns to ds and y
sun_power_data.rename(columns = {'Date' : 'ds', 'Solar Irradiance (kWh/m^2/day)' : 'y'}, inplace = True)

# instantiate and fit prophet
prophet = Prophet()
prophet.fit(sun_power_data)

# make future predictions
future = prophet.make_future_dataframe(periods = 1095) # make future predictions for 3 years

# make forecast
forecast = prophet.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()

# plot the forecast
fig1 = prophet.plot(forecast)
```

```python
# get the trends of the sun power data
fig2 = prophet.plot_components(forecast)
```

```python
# install additional libraries from Prophet
from prophet.plot import plot_plotly, plot_components_plotly

# get the plot_plots
fig3 = plot_plotly(prophet, forecast)
fig3
```
